```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.impute import SimpleImputer
```

```python
dataset = pd.read_csv("loan.csv")
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset.info()
```

```python
dataset.select_dtypes(include="float64")
```

```python
si = SimpleImputer(strategy="mean") #using mean strategy of simpleimputer from scikitlearn
ar = si.fit_transform(dataset[dataset.select_dtypes(include="float64").columns]) #fit_transform fills all the null values using the strategy used
```

```python
#as si.fit_transform gives array result , we convert it into table
new_dataset = pd.DataFrame(ar,columns=dataset.select_dtypes(include="float64").columns) 
```

```python
#check the result
new_dataset.isnull().sum()
```
