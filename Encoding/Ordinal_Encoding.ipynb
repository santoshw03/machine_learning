```python
import pandas as pd
```

```python
#lets take custom dataset
df = pd.DataFrame({"Size":["s","m","l","xl","s","xl","m","m","l","l","s"]})
df.head()
```

```python
# define the order of data:
ord_data = [["s","m","l","xl"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories = ord_data) #telling that we want to encode in this order , i.e order of ord_data
oe.fit(df[["Size"]])
```

```python
oe.transform(df[["Size"]])
```

```python
df["en_size"] = oe.transform(df[["Size"]])
df
```

```python
# thus the ordinal encoder encoded :
# s-> 0 , m->1 , l->2 , xl->3
```

```python
# doing ordinal encoding using map() , by which we can pass any numerical value to the data as we want 
# so by ordinal encoding lets code the sizes with the size in numbers
```

```python
ord_data2 = {"s":39 , "m":40 , "l":42 ,"xl": 44}
```

```python
df["map_encoded_size"] = df["Size"].map(ord_data2)
```

```python
# lets encode property_area and add a new column tier having values :
# urban -> tier 1
# semiurban -> tier 2
# rural -> tier 3
```

```python
ord_property = {"Urban":1,"Semiurban":2,"Rural":3}
```

```python
dataset["area_tier"] = dataset["Property_Area"].map(ord_property)
dataset[["Property_Area","area_tier"]].head(10)
```
