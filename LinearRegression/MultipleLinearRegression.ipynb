```python
# equation for simple linear regression:
# y = mx + c 

# equation for multiple linear regression:
# y = m1x1 + m2x2 + m3x3 +......mnxn + c
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv("Salary_Data.csv")
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
# lets see if the dataset follows linearity or not by using pairplot grapg , which will show relations in pairs in the graph
```

```python
sns.pairplot(data=dataset)
plt.show()
```

```python
# hence our data follows linearity , that is one value increases with the increase in other 
```

```python
# lets split the input data and output data:
x_input = dataset.iloc[:,:-1]
y_output = dataset["Salary"]
```

```python
# now lets split the data for training and testing:
x_train,x_test,y_train,y_test = train_test_split(x_input,y_output,test_size=0.2,random_state=42)
```

```python

# now lets build the model:
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
#training
lr.fit(x_train , y_train)
```

```python
# lets see the accuracy of the trained model:
lr.score(x_test,y_test)*100
```

```python
# hence our model is 88% accurate
```

```python
# lets predict the salary of a person having 3 years experience and age is 25
lr.predict([[3,25]])
```

```python
# hence a 25 year old having 3 years experience can have a salary of 58221 rs
```
